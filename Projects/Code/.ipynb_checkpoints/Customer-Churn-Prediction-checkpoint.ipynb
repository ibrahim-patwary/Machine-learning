# Preprocessing Data

```python
df = df.drop("customerID", axis =1)
```

```python
df.shape
```

# Encoding 

```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()
```

```python
for column in df.columns:
    if df[column].dtype==np.number:
        continue
    df[column] = le.fit_transform(df[column])
      
```

```python
df
```

# Scaling the data set

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['MonthlyCharges']])
df['MonthlyCharges'] = scaler.transform(df[['MonthlyCharges']])

scaler.fit(df[['TotalCharges']])
df['TotalCharges'] = scaler.transform(df[['TotalCharges']])

scaler.fit(df[['tenure']])
df['tenure'] = scaler.transform(df[['tenure']])
```

```python
df.head(10)
```

# Find total Memory Uses

```python
memory = df.memory_usage()
print(memory)
print("Total Memory Usage = ",sum(memory))
```

# Memory Optimzations

```python
df.iloc[:,0:20] = df.iloc[:,0:20].astype('float16')
```

```python
memory = df.memory_usage()
print(memory)
print("Total Memory Usage = ",sum(memory))
```
