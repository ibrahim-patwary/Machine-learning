# For Feature Selection

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import f_classif
```

```python
FIT_FEATURES = SelectKBest(score_func= f_classif)
```

```python
FIT_FEATURES.fit(x,y)
```

```python
SCORE_COL=pd.DataFrame(FIT_FEATURES.scores_,columns =["score value"])
```

```python
SCORE_COL
```

```python
NAME_COL = pd.DataFrame(x.columns)
```

```python
top_feature = pd.concat([NAME_COL, SCORE_COL], axis= 1)
```

```python
top_feature
```

```python
top = top_feature.nlargest(20,"score value")
```

```python
top
```

```python
df = pd.DataFrame(top)
```

```python
df
```

```python
final =df.rename(columns={0: 'Name', 'score value': 'Score'})
```

```python
final
```

```python
x =final["Name"]
y = final["Score"]
```

```python
plt.figure(figsize =(25,10))
plt.bar(x,y, color="hotpink")
plt.xlabel("Features Name")
plt.ylabel("Score")
plt.title("Feature Selection")
plt.savefig("feature_selection.jpeg")
plt.show()
```
