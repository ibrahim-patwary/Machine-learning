```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv("E:\jupyter notebook\Machine Learning\Encoding Techniques\One Hot Encoding\Data set\carprices.csv")
df
```

```python
dummies= pd.get_dummies(df["Car Model"])
```

```python
dummies
```

```python
marged =pd.concat([df,dummies], axis ="columns")
```

```python
marged
```

```python
final = marged.drop(["Car Model"] , axis="columns")
```

```python
final
```

```python
final.isnull().sum()
```

```python
y=final["Sell Price"]
```

```python
y
```

```python
x =final.drop(["Sell Price"] ,axis =1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

# For Audi

```python
model.predict([[45000,8,1,0,0,0]])
```

```python
model.predict([[45000,8,0,0,0,1]])
```
