```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("E:\jupyter notebook\Machine Learning\Encoding Techniques\Ordinal Encoding\Data set\shop data.csv")
df
```

```python
df.isnull().sum()
```

```python
df["income"].unique()
```

```python
income_order =["low","medium","high"]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
model = OrdinalEncoder(categories =[income_order])
```

```python
model.fit(df[["income"]])
```

```python
final = pd.DataFrame(model.transform(df[["income"]]))
```

```python
df["income"]= final
```
