```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("E:\jupyter notebook\Machine Learning\Dicision Tree\Data set\salaries.csv")
df.head()
```

```python
df["company"].unique()
```

```python
inputs = df.drop("salary_more_then_100k" , axis ="columns")
inputs.head()
```

```python
target = df["salary_more_then_100k"]
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs["company_n"] = le_company.fit_transform(inputs["company"])
inputs["job_n"] = le_job.fit_transform(inputs["job"])
inputs["degree_n"] = le_degree.fit_transform(inputs["degree"])
```

```python
inputs
```

```python
inputs.drop(["company","job","degree"] ,axis ="columns", inplace =True )
```

```python
inputs
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(inputs , target)
```

```python
model.score(inputs , target)
```

# For Google =2 ,Job = CSE, Degree = Bsc 

```python
model. predict([[2,1,0]])
```

# For Google =2 ,Job = CSE, Degree = Msc

```python
model. predict([[2,1,1]])
```
